```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from torchvision import datasets, transforms, models
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import shutil

# Install dependencies
!pip install seaborn opencv-python kaggle

# Upload Kaggle API key
from google.colab import files
files.upload()  # Upload kaggle.json

# Set up Kaggle API and download dataset
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json
!rm -rf /content/chest_xray
!kaggle datasets download -d paultimothymooney/chest-xray-pneumonia
!unzip -o -q chest-xray-pneumonia.zip
!rm -rf /content/chest_xray/__MACOSX

# Verify GPU
print(f"CUDA disponible : {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Nom du GPU : {torch.cuda.get_device_name(0)}")

# Set dataset path
data_dir = '/content/chest_xray/chest_xray'
splits = ['train', 'val', 'test']
categories = ['NORMAL', 'PNEUMONIA']

# Verify dataset
print("VÃ©rification des dossiers dans", data_dir)
if not os.path.exists(data_dir):
    print(f"Erreur : Le dossier {data_dir} n'existe pas.")
else:
    !ls -R {data_dir}

# Verify training directory contents
train_dir = os.path.join(data_dir, 'train')
for category in categories:
    folder = os.path.join(train_dir, category)
    images = glob.glob(os.path.join(folder, '*.[jJ][pP][eE][gG]')) + glob.glob(os.path.join(folder, '*.jpg'))
    print(f"Category: {category}, Number of images: {len(images)}")
    if images:
        print(f"Sample image: {images[0]}")
    else:
        print(f"WARNING: No images found in {folder}")

# Step 1: Explore dataset
def explore_dataset():
    data_counts = []
    for split in splits:
        for category in categories:
            folder = os.path.join(data_dir, split, category)
            image_paths = glob.glob(os.path.join(folder, '*.[jJ][pP][eE][gG]')) + glob.glob(os.path.join(folder, '*.jpg'))
            num_images = len(image_paths)
            data_counts.append({'Split': split, 'Category': category, 'Count': num_images})

    df = pd.DataFrame(data_counts)
    print("RÃ©partition des images :")
    print(df.pivot(index='Split', columns='Category', values='Count'))

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Split', y='Count', hue='Category', data=df)
    plt.title('RÃ©partition des images par catÃ©gorie et split')
    plt.savefig('/content/dataset_distribution.png')
    plt.show()

    return df

# Step 2: Check image sizes
def check_image_sizes():
    sizes = []
    for split in splits:
        for category in categories:
            folder = os.path.join(data_dir, split, category)
            image_paths = glob.glob(os.path.join(folder, '*.[jJ][pP][eE][gG]')) + glob.glob(os.path.join(folder, '*.jpg'))
            if not image_paths:
                print(f"Aucune image trouvÃ©e dans : {folder}")
                continue
            for img_path in image_paths:
                try:
                    with Image.open(img_path) as img:
                        sizes.append({'Split': split, 'Category': category, 'Size': img.size})
                except Exception as e:
                    print(f"Erreur lors de la lecture de l'image {img_path} : {e}")

    if not sizes:
        raise ValueError("Aucune image valide n'a Ã©tÃ© trouvÃ©e dans le dataset.")

    df_sizes = pd.DataFrame(sizes)
    print("\nTailles uniques des images :")
    print(df_sizes['Size'].value_counts())
    return df_sizes

# Step 3: Compute class weights
def compute_class_weights(df_counts):
    train_counts = df_counts[df_counts['Split'] == 'train']
    normal_count = train_counts[train_counts['Category'] == 'NORMAL']['Count'].iloc[0]
    pneumonia_count = train_counts[train_counts['Category'] == 'PNEUMONIA']['Count'].iloc[0]

    total = normal_count + pneumonia_count
    if normal_count == 0 or pneumonia_count == 0:
        print("Avertissement : Une classe est vide.")
        return torch.tensor([1.0, 1.0]).float()
    weight_normal = total / (2 * normal_count)
    weight_pneumonia = total / (2 * pneumonia_count)

    return torch.tensor([weight_normal, weight_pneumonia]).float()

# Step 4: Split train/validation
def split_train_val(data_dir, val_ratio=0.1):
    train_dir = os.path.join(data_dir, 'train')
    val_dir = os.path.join(data_dir, 'val_new')
    os.makedirs(val_dir, exist_ok=True)

    for category in ['NORMAL', 'PNEUMONIA']:
        os.makedirs(os.path.join(val_dir, category), exist_ok=True)
        images = glob.glob(os.path.join(train_dir, category, '*.[jJ][pP][eE][gG]')) + glob.glob(os.path.join(train_dir, category, '*.jpg'))
        if not images:
            print(f"Aucune image trouvÃ©e pour {category} dans {train_dir}")
            continue
        train_imgs, val_imgs = train_test_split(images, test_size=val_ratio, random_state=42)

        for img in val_imgs:
            try:
                shutil.move(img, os.path.join(val_dir, category, os.path.basename(img)))
            except Exception as e:
                print(f"Erreur lors du dÃ©placement de {img} : {e}")

    print("Nouveau validation set crÃ©Ã©.")

# Run split_train_val and check for images
try:
    split_train_val(data_dir, val_ratio=0.1)
except Exception as e:
    print(f"Erreur lors de l'exÃ©cution de split_train_val : {e}")

# Verify train directory after split
for category in categories:
    folder = os.path.join(data_dir, 'train', category)
    images = glob.glob(os.path.join(folder, '*.[jJ][pP][eE][gG]')) + glob.glob(os.path.join(folder, '*.jpg'))
    print(f"Post-split - Category: {category}, Number of images: {len(images)}")

# Step 5: Define transformations
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(degrees=15),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.8, 1.2)),
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3),
        transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ]),
    'test': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
}

# Step 6: Custom dataset
class CustomImageDataset(Dataset):
    def __init__(self, image_folder_dataset, transform=None):
        self.image_folder_dataset = image_folder_dataset
        self.transform = transform

    def __len__(self):
        return len(self.image_folder_dataset)

    def __getitem__(self, idx):
        try:
            img, label = self.image_folder_dataset[idx]
            return img, label
        except:
            return self.__getitem__((idx + 1) % len(self))

# Step 7: Load datasets
try:
    image_datasets = {
        'train': datasets.ImageFolder(os.path.join(data_dir, 'train'), data_transforms['train']),
        'val': datasets.ImageFolder(os.path.join(data_dir, 'val_new'), data_transforms['val']),
        'test': datasets.ImageFolder(os.path.join(data_dir, 'test'), data_transforms['test'])
    }
except Exception as e:
    print(f"Erreur lors du chargement des datasets : {e}")
    raise

image_datasets['train'] = CustomImageDataset(image_datasets['train'], data_transforms['train'])
image_datasets['val'] = CustomImageDataset(image_datasets['val'], data_transforms['val'])
image_datasets['test'] = CustomImageDataset(image_datasets['test'], data_transforms['test'])

# Step 8: Create DataLoaders
def get_sampler(dataset):
    targets = [label for _, label in dataset.image_folder_dataset]
    if not targets:
        raise ValueError("Aucune Ã©tiquette trouvÃ©e dans le dataset.")
    class_counts = np.bincount(targets)
    num_samples = len(targets)
    weights = [1.0 / class_counts[label] for label in targets]
    sampler = WeightedRandomSampler(weights, num_samples, replacement=True)
    return sampler

dataloaders = {
    'train': DataLoader(image_datasets['train'], batch_size=32, sampler=get_sampler(image_datasets['train']), num_workers=4),
    'val': DataLoader(image_datasets['val'], batch_size=32, shuffle=False, num_workers=4),
    'test': DataLoader(image_datasets['test'], batch_size=32, shuffle=False, num_workers=4)
}

dataset_sizes = {x: len(image_datasets[x]) for x in ['train', 'val', 'test']}
class_names = image_datasets['test'].image_folder_dataset.classes
print("Tailles des datasets :", dataset_sizes)
print("Noms des classes :", class_names)

# Step 9: Compute class weights
df_counts = explore_dataset()
check_image_sizes()
class_weights = compute_class_weights(df_counts)
criterion = nn.CrossEntropyLoss(weight=class_weights.to(device))

# Step 10: Visualize data
def imshow(img, title=None):
    img = img.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = std * img + mean
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    if title:
        plt.title(title)
    plt.axis('off')

try:
    dataiter = iter(dataloaders['train'])
    images, labels = next(dataiter)
    plt.figure(figsize=(10, 10))
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        imshow(images[i], title=class_names[labels[i]])
    plt.savefig('/content/sample_images.png')
    plt.show()
except Exception as e:
    print(f"Erreur lors de la visualisation des images : {e}")

# Step 11: Configure ResNet-50
try:
    model = models.resnet50(pretrained=True)
except Exception as e:
    print(f"Erreur lors du chargement des poids : {e}")
    model = models.resnet50(pretrained=False)

for param in model.parameters():
    param.requires_grad = False
for param in model.layer3.parameters():
    param.requires_grad = True
for param in model.layer4.parameters():
    param.requires_grad = True
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 2)
model = model.to(device)

optimizer = optim.Adam([
    {'params': model.fc.parameters(), 'lr': 0.001},
    {'params': model.layer3.parameters(), 'lr': 0.0001},
    {'params': model.layer4.parameters(), 'lr': 0.0001}
], lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3, verbose=True)

# Step 12: Training function
def train_model(model, criterion, optimizer, scheduler, num_epochs=25, patience=5):
    best_acc = 0.0
    best_loss = float('inf')
    epochs_no_improve = 0
    early_stop = False

    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []

    for epoch in range(num_epochs):
        if early_stop:
            print("Early stopping triggered.")
            break
        print(f'Epoch {epoch}/{num_epochs - 1}')

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            if phase == 'train':
                train_losses.append(epoch_loss)
                train_accs.append(epoch_acc.item())
            else:
                val_losses.append(epoch_loss)
                val_accs.append(epoch_acc.item())
                scheduler.step(epoch_loss)

                if epoch_acc > best_acc:
                    best_acc = epoch_acc
                    torch.save(model.state_dict(), '/content/best_model.pth')
                if epoch_loss < best_loss:
                    best_loss = epoch_loss
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    if epochs_no_improve >= patience:
                        early_stop = True

    torch.save(model.state_dict(), '/content/final_model.pth')
    print("ModÃ¨le final sauvegardÃ© : /content/final_model.pth")

    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Loss Curves')

    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label='Train Acc')
    plt.plot(val_accs, label='Val Acc')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Accuracy Curves')

    plt.savefig('/content/training_curves.png')
    plt.show()

    return model

# Step 13: Train model
model = train_model(model, criterion, optimizer, scheduler, num_epochs=25)

# Step 14: Evaluate on test set
def evaluate_model(model):
    model.eval()
    running_corrects = 0

    for inputs, labels in dataloaders['test']:
        inputs = inputs.to(device)
        labels = labels.to(device)

        with torch.no_grad():
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

        running_corrects += torch.sum(preds == labels.data)

    test_acc = running_corrects.double() / dataset_sizes['test']
    print(f'Test Accuracy: {test_acc:.4f}')

model.load_state_dict(torch.load('/content/best_model.pth'))
evaluate_model(model)

# Step 15: Detailed metrics
def get_predictions(model, dataloader):
    model.eval()
    all_preds = []
    all_labels = []

    for inputs, labels in dataloader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        with torch.no_grad():
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    return np.array(all_preds), np.array(all_labels)

preds, labels = get_predictions(model, dataloaders['test'])
cm = confusion_matrix(labels, preds)
print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(classification_report(labels, preds, target_names=class_names))

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig('/content/confusion_matrix.png')
plt.show()

# Step 16: Predict and verify single image
def predict_image(image_path, model):
    model.eval()
    image = Image.open(image_path).convert('RGB')
    image = data_transforms['test'](image).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image)
        _, pred = torch.max(outputs, 1)

    return class_names[pred.item()]

def predict_and_display_image(image_path, model, transform, class_names, true_label):
    model.eval()
    image = Image.open(image_path).convert('RGB')
    image_tensor = transform(image).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image_tensor)
        probs = torch.softmax(outputs, dim=1)
        pred = torch.max(outputs, 1)[1]
        pred_prob = probs[0][pred].item()

    pred_label = class_names[pred.item()]
    is_correct = pred_label == true_label

    plt.figure(figsize=(6, 6))
    img = transform(image).cpu()
    title = (f'PrÃ©diction: {pred_label} (Prob: {pred_prob:.4f})\n'
             f'Vrai: {true_label}\n'
             f'Correct: {"Oui" if is_correct else "Non"}')
    imshow(img, title=title)
    plt.savefig('/content/prediction_result.png')
    plt.show()

    print(f'Image: {os.path.basename(image_path)}')
    print(f'PrÃ©diction: {pred_label} (ProbabilitÃ©: {pred_prob:.4f})')
    print(f'Ãtiquette rÃ©elle: {true_label}')
    print(f'PrÃ©diction correcte: {"Oui" if is_correct else "Non"}')

    return pred_label, is_correct

image_path = os.path.join(data_dir, 'test', 'NORMAL', 'IM-0001-0001.jpeg')
true_label = 'NORMAL'

try:
    model.load_state_dict(torch.load('/content/final_model.pth'))
    print("ModÃ¨le final chargÃ© : /content/final_model.pth")
except FileNotFoundError:
    print("Fichier final_model.pth introuvable. Chargement de best_model.pth.")
    model.load_state_dict(torch.load('/content/best_model.pth'))

if os.path.exists(image_path):
    pred_label, is_correct = predict_and_display_image(
        image_path, model, data_transforms['test'], class_names, true_label
    )
else:
    print(f"Image {image_path} introuvable.")

# Step 17: Generate submission file
test_files = glob.glob(os.path.join(data_dir, 'test', '*/*.[jJ][pP][eE][gG]')) + glob.glob(os.path.join(data_dir, 'test', '*/*.jpg'))
submission = []

for img_path in test_files:
    pred = predict_image(img_path, model)
    img_name = os.path.basename(img_path)
    submission.append({'Image': img_name, 'Prediction': pred})

submission_df = pd.DataFrame(submission)
submission_df.to_csv('/content/submission.csv', index=False)
print("Fichier de soumission crÃ©Ã© : /content/submission.csv")
```

```python
import os
import glob
import numpy as np
import cv2
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import shutil
from PIL import Image

# Install dependencies
!pip install opencv-python kaggle

# Upload Kaggle API key
from google.colab import files
files.upload()  # Upload kaggle.json

# Set up Kaggle API and download dataset
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json
!rm -rf /content/chest_xray
!kaggle datasets download -d paultimothymooney/chest-xray-pneumonia
!unzip -o -q chest-xray-pneumonia.zip
!rm -rf /content/chest_xray/__MACOSX

# Verify GPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Utilisation du device : {device}")

# Set dataset path
data_dir = '/content/chest_xray/chest_xray'
categories = ['NORMAL', 'PNEUMONIA']

# Verify dataset
print("VÃ©rification des dossiers dans", data_dir)
if not os.path.exists(data_dir):
    print(f"Erreur : Le dossier {data_dir} n'existe pas.")
else:
    !ls -R {data_dir}

# Generate synthetic masks for PNEUMONIA images
def generate_synthetic_mask(image_path):
    try:
        # Load image in grayscale
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Erreur : Impossible de lire l'image {image_path}")
            return np.zeros((224, 224))  # Return zero mask on failure

        # Apply adaptive thresholding
        thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                      cv2.THRESH_BINARY_INV, 11, 2)

        # Morphological operations
        kernel = np.ones((5,5), np.uint8)
        thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
        thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

        # Normalize mask to [0, 1]
        mask = thresh / 255.0
        return mask
    except Exception as e:
        print(f"Erreur lors de la gÃ©nÃ©ration du masque pour {image_path} : {e}")
        return np.zeros((224, 224))  # Return zero mask on error

# Custom dataset for segmentation
class PneumoniaSegmentationDataset(Dataset):
    def __init__(self, image_paths, transform=None, mask_transform=None):
        self.image_paths = image_paths
        self.transform = transform
        self.mask_transform = mask_transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        try:
            img = Image.open(img_path).convert('RGB')
        except Exception as e:
            print(f"Erreur lors de la lecture de l'image {img_path} : {e}")
            # Return a dummy image and mask to avoid crashing
            img = Image.fromarray(np.zeros((224, 224, 3), dtype=np.uint8))

        # Generate mask
        mask = generate_synthetic_mask(img_path)
        mask = Image.fromarray((mask * 255).astype(np.uint8))

        if self.transform:
            img = self.transform(img)
        if self.mask_transform:
            mask = self.mask_transform(mask)

        return img, mask.float()

# U-Net model
class UNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super(UNet, self).__init__()

        def conv_block(in_channels, out_channels):
            return nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 3, padding=1),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_channels, out_channels, 3, padding=1),
                nn.ReLU(inplace=True)
            )

        def upconv_block(in_channels, out_channels):
            return nn.Sequential(
                nn.ConvTranspose2d(in_channels, out_channels, 2, stride=2),
                nn.ReLU(inplace=True)
            )

        # Encoder
        self.enc1 = conv_block(in_channels, 64)
        self.enc2 = conv_block(64, 128)
        self.enc3 = conv_block(128, 256)
        self.enc4 = conv_block(256, 512)
        self.pool = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = conv_block(512, 1024)

        # Decoder
        self.upconv4 = upconv_block(1024, 512)
        self.dec4 = conv_block(1024, 512)
        self.upconv3 = upconv_block(512, 256)
        self.dec3 = conv_block(512, 256)
        self.upconv2 = upconv_block(256, 128)
        self.dec2 = conv_block(256, 128)
        self.upconv1 = upconv_block(128, 64)
        self.dec1 = conv_block(128, 64)

        # Output
        self.out_conv = nn.Conv2d(64, out_channels, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))
        e3 = self.enc3(self.pool(e2))
        e4 = self.enc4(self.pool(e3))

        # Bottleneck
        b = self.bottleneck(self.pool(e4))

        # Decoder
        d4 = self.upconv4(b)
        d4 = torch.cat([d4, e4], dim=1)
        d4 = self.dec4(d4)
        d3 = self.upconv3(d4)
        d3 = torch.cat([d3, e3], dim=1)
        d3 = self.dec3(d3)
        d2 = self.upconv2(d3)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)
        d1 = self.upconv1(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.dec1(d1)

        out = self.out_conv(d1)
        out = self.sigmoid(out)
        return out

# Dice loss
class DiceBCELoss(nn.Module):
    def __init__(self):
        super(DiceBCELoss, self).__init__()
        self.bce = nn.BCELoss()

    def forward(self, pred, target):
        smooth = 1e-6
        pred_flat = pred.view(-1)
        target_flat = target.view(-1)
        intersection = (pred_flat * target_flat).sum()
        dice = (2. * intersection + smooth) / (pred_flat.sum() + target_flat.sum() + smooth)
        dice_loss = 1 - dice
        bce_loss = self.bce(pred, target)
        return bce_loss + dice_loss

# Prepare dataset
pneumonia_images = glob.glob(os.path.join(data_dir, 'train/PNEUMONIA', '*.[jJ][pP][eE][gG]')) + \
                   glob.glob(os.path.join(data_dir, 'train/PNEUMONIA', '*.jpg'))
print(f"Nombre d'images PNEUMONIA : {len(pneumonia_images)}")

if not pneumonia_images:
    raise ValueError("Aucune image PNEUMONIA trouvÃ©e. VÃ©rifiez le dossier du dataset.")

# Split into train and validation
train_imgs, val_imgs = train_test_split(pneumonia_images, test_size=0.2, random_state=42)

# Define transformations
image_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

mask_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Create datasets
train_dataset = PneumoniaSegmentationDataset(train_imgs, image_transform, mask_transform)
val_dataset = PneumoniaSegmentationDataset(val_imgs, image_transform, mask_transform)

# Create dataloaders with reduced num_workers
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, num_workers=2)

# Initialize model, loss, and optimizer
model = UNet(in_channels=3, out_channels=1).to(device)
criterion = DiceBCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3)

# Training function
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=20, patience=5):
    best_loss = float('inf')
    epochs_no_improve = 0
    train_losses = []
    val_losses = []

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        for images, masks in train_loader:
            images, masks = images.to(device), masks.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, masks)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * images.size(0)

        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for images, masks in val_loader:
                images, masks = images.to(device), masks.to(device)
                outputs = model(images)
                loss = criterion(outputs, masks)
                val_loss += loss.item() * images.size(0)

        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        scheduler.step(val_loss)

        print(f'Epoch {epoch}/{num_epochs-1}')
        print(f'Train Loss: {train_loss:.4f}')
        print(f'Val Loss: {val_loss:.4f}')

        if val_loss < best_loss:
            best_loss = val_loss
            torch.save(model.state_dict(), '/content/best_unet.pth')
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print("Early stopping triggered.")
                break

    # Plot loss curves
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Loss Curves')
    plt.savefig('/content/unet_loss_curves.png')
    plt.show()

    return model

# Train the model
try:
    model = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs=20)
except Exception as e:
    print(f"Erreur lors de l'entraÃ®nement : {e}")

# Evaluate and visualize
def visualize_segmentation(model, image_path, transform, true_label='PNEUMONIA'):
    model.eval()
    try:
        img = Image.open(image_path).convert('RGB')
    except Exception as e:
        print(f"Erreur lors de la lecture de l'image {image_path} : {e}")
        return

    img_tensor = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        pred_mask = model(img_tensor).squeeze().cpu().numpy()

    pred_mask = (pred_mask > 0.5).astype(np.uint8)  # Threshold to binary mask

    # Load original image for visualization
    img_np = np.array(img.resize((224, 224)))

    # Plot
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(img_np)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(pred_mask, cmap='gray')
    plt.title('Predicted Mask')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(img_np)
    plt.imshow(pred_mask, cmap='jet', alpha=0.5)
    plt.title('Overlay')
    plt.axis('off')

    plt.savefig('/content/segmentation_result.png')
    plt.show()

    print(f'Image: {os.path.basename(image_path)}')
    print(f'Ãtiquette rÃ©elle: {true_label}')

# Test segmentation on a PNEUMONIA image
image_path = os.path.join(data_dir, 'test/PNEUMONIA', 'person1_virus_6.jpeg')
if os.path.exists(image_path):
    visualize_segmentation(model, image_path, image_transform, true_label='PNEUMONIA')
else:
    print(f"Image {image_path} introuvable.")
```
